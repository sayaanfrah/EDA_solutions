# Understanding the Dataset 

There are 9 columns in the dataset mainly :

1. Cement (component 1)(kg in a m^3 mixture)
2. Blast Furnace Slag (component 2)(kg in a m^3 mixture)
3. Fly Ash (component 3)(kg in a m^3 mixture)
4.  Water  (component 4)(kg in a m^3 mixture)
5. Superplasticizer (component 5)(kg in a m^3 mixture)
6. Coarse Aggregate  (component 6)(kg in a m^3 mixture)
7. Fine Aggregate (component 7)(kg in a m^3 mixture) 
8. Age (day)
9. Concrete compressive strength(MPa, megapascals)


Cement (component 1)(kg in a m^3 mixture) : It shows the density of the cement Since the unit of the column is kg/m^3.The unit of density is also Kilogram per cubic meter.

Blast Furnace Slag (component 2)(kg in a m^3 mixture): It shows the density of the slag formed in the blast furnace.This is basically the residue which is left  in the furnace after the heating .it remains in the fluid form .

Fly Ash (component 3)(kg in a m^3 mixture):It shows the density of the Ash left after heating the furnace .

Water  (component 4)(kg in a m^3 mixture):It shows the density of the water required during the burninin og the furnace.

Superplasticizer (component 5)(kg in a m^3 mixture): It is also known as high range water reducers ,are additives used in making high strength conerete.The Column shows its density

Coarse Aggregate  (component 6)(kg in a m^3 mixture):it is Construction aggregate, or simply aggregate, is a broad category of coarse- to medium-grained particulate material used in construction, including sand, gravel, crushed stone, slag, recycled concrete and geosynthetic aggregates. Aggregates are the most mined materials in the world. This column shows its density

Fine Aggregate (component 7)(kg in a m^3 mixture):It shows the density of Fine Aggregate used in blast furnace

Age (day):This column gives the idea of how old the suppliers are .
 
Concrete compressive strength(MPa, megapascals): Concrete compressive strength is basically the load per unit area.It also gives the yielding strength and life of the concrete used in its building .

# Removing skewness by transforming Data

```python
from sklearn.preprocessing import power_transform
x=power_transform(x,method="yeo-johnson")
```

# Regressor Model Building

# Linear Regression:

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_t=sc.fit_transform(x)
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
max_scr=0
for i in range(0,1000):
    train_x,test_x,train_y,test_y=train_test_split(x_t,y,random_state=i,test_size=0.20)
    lr.fit(train_x,train_y)
    pred_train=lr.predict(train_x)
    pred_test=lr.predict(test_x)
    if round(r2_score(train_y,pred_train)*100,2)==round(r2_score(test_y,pred_test)*100,2):
        print("At Random State",i,"The model performs very well")
        print("At Random State i:-",i)
        print("Training r2_score is ",r2_score(train_y,pred_train)*100)
        print("Testing r2_score is  ",r2_score(test_y,pred_test)*100)
```

```python
x_train_b,x_test_b,y_train_b,y_test_b=train_test_split(x_t,y,random_state=96,test_size=0.20)
```

We will use Random State 96

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
import warnings
warnings.filterwarnings("ignore")
```

# Regularization

# 1. Lasso:

```python
from sklearn.linear_model import Lasso
parameters={"alpha":[0.0001,0.001,0.01,0.1,1,10,100],"random_state":list(range(0,10))}
ls=Lasso()
clf=GridSearchCV(ls,parameters)
clf.fit(train_x,train_y)
print(clf.best_params_)
```

```python
ls=Lasso(alpha=0.1,random_state=0)
ls.fit(x_train_b,y_train_b)
ls.score(x_train_b,y_train_b)
pred_ls=ls.predict(x_test_b)

lss=r2_score(y_test_b,pred_ls)
for j in range(2,10):
    lsscore=cross_val_score(ls,x_t,y,cv=j)
    lsc=lsscore.mean()
    print("At cv:-",j)
    print("cross_validation_score is:-",lsc*100)
    print("r2_score is:-",lss*100)
    print("\n")
```

```python
print("At cv:-",3)
print("Cross Validation Score is ",72.366)
print("r2_score is",81)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
print("Error: ")


print("mean_squared_error",mean_squared_error(y_test_b,pred_ls))
print("mean_absolute_error",mean_absolute_error(y_test_b,pred_ls))
print("root mean squared error",np.sqrt(mean_squared_error(y_test_b,pred_ls)))
```

# 2. Ridge Regression

```python
from sklearn.linear_model import Ridge
parameters={"alpha":[.0001,.001,.01,.1,1],"fit_intercept":[True,False],"normalize":[True,False],"copy_X":[True,False],"tol":[.0001,.001,.01,.1,1],"random_state":list(range(0,10))}
rd=Ridge()
clf=GridSearchCV(rd,parameters)
clf.fit(x_train_b,y_train_b)

print(clf.best_params_)
```

```python
rd=Ridge(alpha=0.01,copy_X=True,fit_intercept=True,normalize=True,random_state=0,tol=0.0001)
rd.fit(x_train_b,y_train_b)
rd.score(x_train_b,y_train_b)
pred_rd=rd.predict(x_test_b)

rds=r2_score(y_test_b,pred_rd)
print("R2 Score:",rds*100)

rdscore=cross_val_score(rd,x_t,y,cv=3)
rdc=rdscore.mean()
print("Cross val score",rdc*100)

```

```python
print("Error:")


print("mean_squared_error",mean_squared_error(y_test_b,pred_rd))
print("mean_aboslute_error",mean_absolute_error(y_test_b,pred_rd))
print("root mean squared error",np.sqrt(mean_squared_error(y_test_b,pred_rd)))
```

```python
plt.figure(figsize=(10,7))
plt.scatter(x=y_test_b,y=pred_rd,color='r')
plt.plot(y_test_b,y_test_b,color='lime')
plt.xlabel("Actual Compressive Strength",fontsize=20)
plt.ylabel("Predictive Compressive Strength",fontsize=20)
plt.title("Ridge Regression",fontsize=34)
plt.show()
```
